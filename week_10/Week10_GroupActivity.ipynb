One hot encoding is a method to convert categorical data to numerical data. It represents the categorical variables as binary vectors. 

```python
#define the neighbourhood groups 
groups = belize_df['neighbourhood_group'].unique()
print(groups)
#map each neighborhoods to an integer
group_to_int = dict((group, i) for i, group in enumerate(groups))
int_to_group = dict((i, group) for i, group in enumerate(groups))

#integer encode input data
group_encoded_df = [group_to_int[data] for data in belize_df['neighbourhood_group']]
print(group_encoded_df)

onehot_encoded = list()
for value in group_encoded_df:
    #creating a matrix of 6 since we have 6 neighborhoods
    neighborhood = [0 for _ in range(len(groups))]
    neighborhood[value] = 1
    onehot_encoded.append(neighborhood)
print(onehot_encoded)
```

```python
inverted = int_to_group[np.argmax(onehot_encoded[0])]
print(inverted)
```

```python
# Convert encoded list into a series
s = pd.Series(onehot_encoded)
# Merge encoded series back into df
belize_df2 = pd.concat([belize_df, s], axis=1)
belize_df2.head()
```

# 1e. Which month (over the years) has had the most shows?

```python
jeopardy_df.groupby(' Air Date')['Show Number'].agg(pd.Series.drop_duplicates)

jeopardy_df[' Air Date'] = pd.to_datetime(jeopardy_df[' Air Date'])

Shows_Month = pd.DataFrame(jeopardy_df.groupby(jeopardy_df[' Air Date'].dt.month)['Show Number'].nunique())
Shows_Month.columns = ['Number of Shows']
Shows_Month

Shows_Month.sort_values('Number of Shows', ascending=False)

#November was the month with the most shows.
```

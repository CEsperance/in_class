# 2. Create a function that accepts a list of values. Filter out all values that are numeric and return the new list.

Input:

[1,2,'a','b',7.6]

output: 

[‘a’,’b’]

```python
l_list = [1,2,'a','b',7.6]

def only_letters(l):
    #create an empty list to add only the letters to
    new_list = []
    #if the item (called letter) in the list is a string, add it to new_list
    for letter in l:
        if type(letter) == str:
            new_list.append(letter)
    return new_list

only_letters(l_list)
```

## Lambda
- Lambda functions is an anonymous function which means, it doesn't have a name
- have less code than a normal function

Use lambda as function (starts with def), to filter results, apply results to a column or row, call a more complicated function, etc.

```python
jeopardy_df = pd.read_csv("JEOPARDY_CSV.csv")
jeopardy_df.head()
```

```python
jeopardy_df.columns = jeopardy_df.columns.str.replace(" ", "")
jeopardy_df.head()
```

```python
jeopardy_df.dtypes
```

```python
jeopardy_df['AirDate']=pd.to_datetime(jeopardy_df['AirDate'], format='%m/%d/%Y')
jeopardy_df
```

```python
jeopardy_df = jeopardy_df.assign(month=lambda x: x['AirDate'].dt.month)
jeopardy_df.head()
```

```python
jeopardy_df['month2'] = jeopardy_df.apply(month=lambda x: x['AirDate'].dt.month, axis=1)
jeopardy_df.head()
```

```python
# clean the data:  $, commas, spaces
# can ither inlcue all logic in the  lambda , or replace the characters we dont want adn then run the lambda to filter
clean_up_lambda=lambda num:int(num.replace('$','').replace(' ','').replace(',',''))

filtered_list = list(filter(lambda num:int(num.replace('$','').replace(' ','').replace(',',''))>300, value_list))
filtered_list
```

# Map and Reduce

Map allows us to apply a function over all the input values. It allows use to iterate over various sizes of inputs: whole dataframe, column, list

Reduce performs a computation or aggregation over the whole list.

```python
mapped_list = list(map(lambda num:int(num.replace('$','').replace(' ','').replace(',',''))>300, value_list))
mapped_list
```
